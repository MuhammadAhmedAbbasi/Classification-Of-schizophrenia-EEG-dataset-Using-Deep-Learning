```python
import mne
import numpy as np
from sklearn import preprocessing
from os.path import exists
```

```python
file_name =['h','s']
h_num=14
s_num=14
iir_params = {'order': 2, 'ftype': 'butter'}
```

```python
def select_channel(raw,selected_channel=['Fp1','Fp2']):
    chan_indices=[]
    for channel in selected_channel:
        chan_indices.append(raw.ch_names.index(channel))
    new_raw=raw.get_data(picks=chan_indices)
    return new_raw
```

```python
def division(new_raw,label,samp_freq=250,recording_time=10):
    trail_length=samp_freq*recording_time   # Calculating Length of each Trail
    trail_length_shape=np.shape(new_raw)[1]   # Determining the time of each channel
    total_chunks=trail_length_shape// trail_length    # Final How many chunks can be created 
    new_raw=new_raw[:,0:total_chunks*trail_length]     # Changing Shape of Data
    chunks_raw=np.split(new_raw,total_chunks,axis=1)    # Creating the trails
    features=[]
    for chunk in chunks_raw:
        features.append((chunk,label))
    return features


```

```python
class_1=[]
class_2=[]
for names in file_name:
    if names=='h':
        for i in range (1,h_num):
                if i<10:
                    if exists('D:\Schzo\Dataset\h0{}.edf'.format(i)):
                        file='D:\Schzo\Dataset\h0{}.edf'.format(i)
                        raw=mne.io.read_raw_edf(file,eog=None,preload=True,verbose=0,misc=None,stim_channel='auto',exclude=())
                        raw.pick_types(eeg=True)
                        raw.filter(l_freq=0.5, h_freq=50, method='iir', iir_params=iir_params)
                        new_raw=select_channel(raw)
                        chunks=division(new_raw,label=0)
                        class_1+=chunks
                else:
                        if exists('D:\Schzo\Dataset\h{}.edf'.format(i)):
                            file='D:\Schzo\Dataset\h{}.edf'.format(i)
                            raw=mne.io.read_raw_edf(file,eog=None,preload=True,verbose=0,misc=None,stim_channel='auto',exclude=())
                            raw.pick_types(eeg=True)
                            raw.filter(l_freq=0.5, h_freq=50, method='iir', iir_params=iir_params)
                            new_raw=select_channel(raw)
                            chunks=division(new_raw,label=0)
                            class_1+=chunks
    if names=='s':
        for i in range (1,s_num):
                if i<10:
                    if exists('D:\Schzo\Dataset\s0{}.edf'.format(i)):
                        file='D:\Schzo\Dataset\s0{}.edf'.format(i)
                        raw=mne.io.read_raw_edf(file,eog=None,preload=True,verbose=0,misc=None,stim_channel='auto',exclude=())
                        raw.pick_types(eeg=True)
                        raw.filter(l_freq=0.5, h_freq=50, method='iir', iir_params=iir_params)
                        new_raw=select_channel(raw)
                        chunks=division(new_raw,label=1)
                        class_2+=chunks
                else:
                        if exists('D:\Schzo\Dataset\s{}.edf'.format(i)):
                            file='D:\Schzo\Dataset\s{}.edf'.format(i)
                            raw=mne.io.read_raw_edf(file,eog=None,preload=True,verbose=0,misc=None,stim_channel='auto',exclude=())
                            raw.pick_types(eeg=True)
                            raw.filter(l_freq=0.5, h_freq=50, method='iir', iir_params=iir_params)
                            new_raw=select_channel(raw)
                            chunks=division(new_raw,label=1)
                            class_2+=chunks
                            


```
